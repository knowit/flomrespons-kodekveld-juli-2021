```python
import pandas as pd

station_name = 'Brustuen'
base_url = 'https://flomrespons-public-data.s3.eu-central-1.amazonaws.com'
```

```python
# Water level in meter over sea level
waterLevel = pd.read_csv(f'{base_url}/hydmet-dataset/{station_name}/waterLevel.csv')

waterLevel
```

```python
waterLevel.plot(x='time', y='value', figsize=(15,8))
```

```python
# Water discharge in cubic meter per second
waterDischarge = pd.read_csv(f'{base_url}/hydmet-dataset/{station_name}/waterDischarge.csv')

waterDischarge
```

```python
waterDischarge.plot(x='time', y='value', figsize=(15,8))
```

```python
observedWeather = pd.read_csv(f'{base_url}/hydmet-dataset/{station_name}/observedWeather.csv')

observedWeather
```

```python
observedWeather.plot(x='time', y='precipitation_amount', figsize=(15,8))
```

```python
forecastedWeather = pd.read_csv(f'{base_url}/hydmet-dataset/{station_name}/forecastedWeather.csv')

forecastedWeather
```

```python
forecastedWeather[forecastedWeather.reference_time == '2021-06-28T00:00:00Z']\
    .sort_values('time')\
    .plot(x='time', y='air_temperature_2m', figsize=(15,8))
```
